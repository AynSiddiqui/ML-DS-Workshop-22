```python
import pandas as pd
from IPython.display import display

penguin_df = pd.read_csv('./penguins_size.csv')
display(penguin_df.head(10))
display(penguin_df['species'].unique())
```

```python
import seaborn as sns

sns.heatmap(penguin_df.isna(), cbar=False)
```

```python
import matplotlib.pyplot as plt

plt.scatter(penguin_df['flipper_length_mm'], penguin_df['body_mass_g'])
plt.xlabel('flipper length in mm')
plt.ylabel('body mass in g')
```

```python
gentoo_data = penguin_df.loc[penguin_df['species']=='Gentoo']

plt.scatter(gentoo_data['flipper_length_mm'], gentoo_data['body_mass_g'])
plt.xlabel('flipper length in mm')
plt.ylabel('body mass in g')
plt.title('Gentoo mass and flipper length')
```

```python
approx_x=[205,230]
approx_y=[4000, 6000]

plt.plot(approx_x, approx_y)
plt.scatter(gentoo_data['flipper_length_mm'], gentoo_data['body_mass_g'])
plt.xlabel('flipper length in mm')
plt.ylabel('body mass in g')
plt.title('Gentoo mass and flipper length')
```

```python
def diff(data, p1, p2):
    slope=(p2[1]-p1[1])/(p2[0]-p1[0])
    pred_mass=(data['flipper_length_mm']-205)*slope+4000
    mass_diff=data['body_mass_g']-pred_mass
    return mass_diff

fig, axs = plt.subplots(2,1)
axs[0].scatter(gentoo_data['flipper_length_mm'], diff(gentoo_data, (205,4000), (230,6000)))
axs[1].scatter(gentoo_data['flipper_length_mm'], [x if (x>500 or x<-500) else 0 for x in diff(gentoo_data, (205,4000), (230,6000))])
```

```python
adelie_data=penguin_df.loc[penguin_df['species']=='Adelie']

plt.scatter(gentoo_data['flipper_length_mm'], gentoo_data['body_mass_g'])
plt.scatter(adelie_data['flipper_length_mm'], adelie_data['body_mass_g'])
```
